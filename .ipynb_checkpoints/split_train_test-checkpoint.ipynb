```python
from matplotlib import pyplot
from matplotlib.image import imread
# define location of dataset
folder = 'scripts/'
# plot first few images
for i in range(1,17):
	# define subplot
	pyplot.subplot(4,4,i)
	# define filename
	filename = folder + 'script' + str(i) + '.png'
	# load image pixels
	image = imread(filename)
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
pyplot.show()
```
